```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) #remove punctuation
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
train['comment_text'] = train['comment_text'].apply(lambda x: clean_text(x))
test['comment_text'] = test['comment_text'].apply(lambda x: clean_text(x))
```

```python
import re
import numpy as np
import torch as th
import torch.autograd as ag
import torch.nn.functional as F
import torch.nn as nn

th.manual_seed(1) # set the seed 


def clean_str(string, tolower=True):
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"\'s", " \'s", string)
    string = re.sub(r"\'ve", " \'ve", string)
    string = re.sub(r"n\'t", " n\'t", string)
    string = re.sub(r"\'re", " \'re", string)
    string = re.sub(r"\'d", " \'d", string)
    string = re.sub(r"\'ll", " \'ll", string)
    string = re.sub(r",", " , ", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", " \( ", string)
    string = re.sub(r"\)", " \) ", string)
    string = re.sub(r"\?", " \? ", string)
    string = re.sub(r"\s{2,}", " ", string)
    if tolower:
        string = string.lower()
    return string.strip()


def loadTexts(l, limit):
    dataset=[]
    for i in range(limit):
        line=l[i]
        cleanline = clean_str(line).split()
        dataset.append(cleanline)
    return dataset
```

```python
LIM=5000 #train.shape[0]
#comment_text=loadTexts(comment_text, LIM)
comment_text_toxic=loadTexts(comment_text_toxic, LIM)
comment_text_unclassified=loadTexts(comment_text_unclassified, LIM)
#comment_text_classified=loadTexts(comment_text_classified, LIM)
```

```python
#comment_text_toxic
```

```python
w2idx = {}

txtidx = []
for sent in comment_text_toxic+comment_text_unclassified : #comment_text_classified
    isent = [] 
    for w in sent: 
        widx=len(w2idx)
        if w in w2idx:
            widx=w2idx[w]         
        else :                    
            w2idx[w]=widx
        isent.append(widx)
    txtidx.append(th.LongTensor(list(set(isent))))
    
print(len(w2idx), "words in the vocab")
print(len(txtidx), "sentences")
print(txtidx[0])

### For the labels
labels = th.ones([2*LIM])
labels[0:LIM] = 1

#labels = train['classified']
```

```python
class CBOW_classifier(nn.Module):

    def __init__(self, vocab_size, embedding_dim):
        super(CBOW_classifier, self).__init__()
        self.embeddings = th.nn.Embedding(
                num_embeddings=vocab_size,
                embedding_dim=embedding_dim)
        self.lin = th.nn.Linear(embedding_dim,1)
        #self.activation = th.nn.Sigmoid()
        
    def forward(self, inputs):
        out = self.embeddings(inputs)
        out = out.sum(dim=0)
        #return self.activation(self.lin(out))
        return self.lin(out)
    
```

```python
classifier = CBOW_classifier(vocab_size=len(w2idx),embedding_dim=10)
```

```python
loss_function = th.nn.BCEWithLogitsLoss()
optimizer = th.optim.SGD(classifier.parameters(),lr=1e-1)
```

```python
import random 
```

```python
total = 2*LIM
nEpoch = 20
randomidx = list(range(total))
preds = th.zeros(total)
losses = th.zeros(nEpoch)
for epoch in range(nEpoch):
    total_loss = th.Tensor([0])
    correct=0 
    random.shuffle(randomidx)
    for i in randomidx:
        x = txtidx[i]
        classifier.zero_grad()
        out = classifier(x)[0]    
        loss = loss_function(out, labels[i])
        pred= out>0
        preds[i] = pred
        if pred.item() == labels[i].item() : 
            correct +=1
        
        loss.backward()
        optimizer.step()
        total_loss += loss.data #on calcule la loss pour chaque sentence
    losses[epoch] = total_loss 
    print(epoch, total_loss[0], 100.0*correct/total, preds.sum())
```
